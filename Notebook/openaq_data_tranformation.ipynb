```python
import pandas as pd

files = [
    "/lakehouse/default/Files/Bronze_openaq/need_openaq_location_7th Ave_measurments_1.csv",
    "/lakehouse/default/Files/Bronze_openaq/need_openaq_location_7th Ave_measurments_2.csv",
]

use_cols = [
    "location_id",
    "location_name",
    "parameter",
    "value",
    "unit",
    "datetimeUtc",
    "latitude",
    "longitude",
]

df = pd.concat(
    [pd.read_csv(f, usecols=use_cols) for f in files],
    ignore_index=True
)

df
```

```python
import numpy as np

# типы
df["location_id"] = pd.to_numeric(df["location_id"], errors="coerce")
df["value"] = pd.to_numeric(df["value"], errors="coerce")
df["ts_utc"] = pd.to_datetime(df["datetimeUtc"], errors="coerce", utc=True)

# фильтры
df = df.dropna(subset=["location_id", "value", "ts_utc"])

df["parameter"] = df["parameter"].astype(str).str.lower().str.strip()
df.loc[df["parameter"].isin(["pm2.5", "pm25"]), "parameter"] = "pm25"
df.loc[df["parameter"].isin(["temperature", "temp"]), "parameter"] = "temperature"

df = df[df["parameter"].isin(["pm25", "temperature"])].copy()

# start = pd.Timestamp("2024-07-01", tz="UTC")
# end   = pd.Timestamp("2024-09-01", tz="UTC")
# df = df[(df["ts_utc"] >= start) & (df["ts_utc"] < end)].copy()

# дата/час
df["date"] = df["ts_utc"].dt.date
df["hour"] = df["ts_utc"].dt.hour

df
```

```python
#Insight 1 — Pollution “hotspots” by time (пики по часам/дням)

# Топ-10 часов с самым высоким PM2.5

pm = df[df["parameter"] == "pm25"].copy()

pm_hourly = (pm.groupby("hour", as_index=False)
               .agg(avg_pm25=("value", "mean"),
                    p95_pm25=("value", lambda s: np.percentile(s, 95)),
                    n=("value", "size"))
               .sort_values("avg_pm25", ascending=False))

pm_hourly.head()
```

```python
# B) Топ-10 дней с самым высоким PM2.5

pm_daily = (pm.groupby("date", as_index=False)
              .agg(avg_pm25=("value", "mean"),
                   max_pm25=("value", "max"),
                   n=("value", "size"))
              .sort_values("avg_pm25", ascending=False))

pm_daily.head(10)
```

```python
# Insight 2 — Seasonal variations (внутри периода)

pm["month"] = pd.to_datetime(pm["date"]).dt.month

pm_month = (pm.groupby("month", as_index=False)
              .agg(avg_pm25=("value", "mean"),
                   median_pm25=("value", "median"),
                   n=("value", "size"))
              .sort_values("month"))

pm_month
```

```python
# Insight 3 — Correlate taxi demand vs PM2.5 (daily)

# A) pm25 + temperature

air_daily = (df.groupby(["date", "parameter"], as_index=False)
               .agg(avg_value=("value", "mean")))

air_wide = air_daily.pivot(index="date", columns="parameter", values="avg_value").reset_index()

air_wide
```

```python
# B) joining open air with taxi data
joined = taxi[taxi["PULocationID"] == 90].merge(air_wide, on="date", how="inner").sort_values("date")

# корреляция trips vs pm25
corr = joined[["trips_count", "pm25"]].corr().iloc[0,1]

# Если хочешь увидеть “в какие дни совпали пики”:

joined["pm25_rank"] = joined["pm25"].rank(ascending=False)
joined["trips_rank"] = joined["trips_count"].rank(ascending=False)

import pandas as pd
joined['date'] = pd.to_datetime(joined['date'])

joined["month"] = joined['date'].dt.month_name()

joined.sort_values(["pm25"], ascending=False)
```
